```python
from transformers import pipeline

# Load a text-generation model using HuggingFace pipeline
generator = pipeline("text-generation", model="gpt2", framework='pt')

# Create a batch of prompts
prompts = [
    "Renewable energy comes from unlimited,",
    "naturally replenished resources, such as the sun, tides, and wind.",
    "Renewable energy can be used for electricity generation,"
    "space and water heating and cooling, and transportation."
]

# Generate text for each prompt
for prompt in prompts:
    output = generator(prompt, max_new_tokens=50, temperature=1.0, num_return_sequences=1)
    print(f"Prompt: {prompt}\nGenerated: {output[0]['generated_text']}\n")
```

```python
# Vary parameters and models
parameters = [
    {"model": "gpt2", "temperature": 0.7, "max_new_tokens": 30},
    {"model": "gpt2", "temperature": 1.5, "max_new_tokens": 50},
    {"model": "EleutherAI/gpt-neo-1.3B", "temperature": 1.0, "max_new_tokens": 40},
    {"model": "microsoft/Phi-3-mini-4k-instruct", "temperature": 0.5, "max_new_tokens": 20},
]

for param in parameters:
    generator = pipeline("text-generation", model=param["model"], framework='pt')
    print(f"\nModel: {param['model']}, Temp: {param['temperature']}, Max Tokens: {param['max_new_tokens']}")
    for prompt in prompts[:2]:  # Test on first two prompts
        output = generator(prompt, max_new_tokens=param["max_new_tokens"], temperature=param["temperature"])
        print(f"Prompt: {prompt}\nGenerated: {output[0]['generated_text']}\n")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from bertviz import model_view

# Load two models of different sizes
small_model_name = "distilgpt2"
large_model_name = "EleutherAI/gpt-neo-1.3B"

# Analyze the smaller model
small_model = AutoModelForCausalLM.from_pretrained(small_model_name, output_attentions=True)
small_tokenizer = AutoTokenizer.from_pretrained(small_model_name)
small_inputs = small_tokenizer.encode("The future of renewable energy is bright", return_tensors="pt")
small_outputs = small_model(small_inputs)
small_attention = small_outputs[-1]
small_tokens = small_tokenizer.convert_ids_to_tokens(small_inputs[0])
# Visualize attention for the small model
print("Small Model Attention:")
model_view(small_attention, small_tokens)

# Analyze the larger model
large_model = AutoModelForCausalLM.from_pretrained(large_model_name, output_attentions=True)
large_tokenizer = AutoTokenizer.from_pretrained(large_model_name)
large_inputs = large_tokenizer.encode("The future of renewable energy is bright", return_tensors="pt")
large_outputs = large_model(large_inputs)
large_attention = large_outputs[-1]
large_tokens = large_tokenizer.convert_ids_to_tokens(large_inputs[0])
# Visualize attention for the large model
print("Large Model Attention:")
model_view(large_attention, large_tokens)
```
